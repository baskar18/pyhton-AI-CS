9.sort an array


```python
a=[22,3,4,55,62]
temp=0
for i in range(0,len(a)):
    for j in range(i+1,len(a)):
        if (a[i]>a[j]):
            temp=a[i]
            a[i]=a[j]
            a[j]=temp
print(a)
```
